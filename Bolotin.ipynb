```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def K_x(tau):
    integrand = lambda w: np.exp(1j * w * tau) / (1 + 2 * w**2 + w**4)
    integral_result, _ = quad(integrand, -np.inf, np.inf)
    return integral_result / (2 * np.pi)

# Создаем массив значений аргумента tau
tau_values = np.linspace(-10, 10, 1000)

# Вычисляем соответствующие значения функции K_x(tau)
K_x_values = [K_x(tau) for tau in tau_values]

# Строим график
plt.plot(tau_values, K_x_values)
plt.title("График функции $K_x(\\tau)$")
plt.xlabel("$\\tau$")
plt.ylabel("$K_x(\\tau)$")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функции
def f1(t):
    return 1/4 * (np.exp(-t) * (t + 1))

def f2(t):
    return -1/4 * (np.exp(t) * (t - 1))

# Создаем массивы значений аргумента
t1 = np.linspace(0, 10, 1000)
t2 = np.linspace(-10, 0, 1000)

# Создаем график
plt.plot(t1, f1(t1), label='f1(t)')
plt.plot(t2, f2(t2), label='f2(t)')

# Настраиваем оси, легенду и сетку
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.grid(True)

# Отображаем график
plt.show()
```
